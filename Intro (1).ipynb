```python
import pandas as pd
import matplotlib.pyplot as plt

# Replace this with the actual path to your CSV file
file_path_line_plot = 'Student_Data.csv'

# Read the data from CSV into a pandas DataFrame
data_line_plot = pd.read_csv(file_path_line_plot)

# Assuming the CSV contains columns 'x' and 'y' for x and y-axis data
x_data = data_line_plot['schoolid']
y_data = data_line_plot['schoolname']

# Plotting the line graph
plt.figure(figsize=(8, 6))
plt.plot(x_data, y_data, marker='o', linestyle='-', color='b')
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

```python
# Replace this with the actual path to your CSV file
file_path_bar_plot = 'Student_Data.csv'

# Read the data from CSV into a pandas DataFrame
data_bar_plot = pd.read_csv(file_path_bar_plot)

# Assuming the CSV contains columns 'categories' and 'values' for bar plot
categories = data_bar_plot['SchoolType']
values = data_bar_plot['TermType']

# Plotting the bar graph
plt.figure(figsize=(8, 6))
plt.bar(categories, values, color='g')
plt.title('Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.xticks(rotation=45)
plt.show()
```

```python
# Replace this with the actual path to your CSV file
file_path_scatter_plot = 'Student_Data.csv'

# Read the data from CSV into a pandas DataFrame
data_scatter_plot = pd.read_csv(file_path_scatter_plot)

# Assuming the CSV contains columns 'x' and 'y' for scatter plot
x_values = data_scatter_plot['schoolid']
y_values = data_scatter_plot['CalendarYear']

# Plotting the scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(x_values, y_values, color='r', alpha=0.5)
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```
